```python
from google.colab import files

# Upload the actual image file
uploaded_actual = files.upload()

# Upload the predicted image file
uploaded_predicted = files.upload()


import tensorflow as tf

# Load the actual image and the predicted image
actual_image = tf.io.read_file("actual_image1.png")
actual_image = tf.image.decode_png(actual_image, channels=3)
predicted_image = tf.io.read_file("predicted_image1.png")
predicted_image = tf.image.decode_png(predicted_image, channels=3)

# Resize the images to the same size
actual_image = tf.image.resize(actual_image, [1024, 1024])
predicted_image = tf.image.resize(predicted_image, [1024, 1024])

# Compute the SSIM between the two images
ssim = tf.image.ssim(actual_image, predicted_image, max_val=255)

print("SSIM:", ssim.numpy())
```
